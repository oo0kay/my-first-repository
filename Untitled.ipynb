```python
#设置切分区域
listBins = [300,500, 550, 600, 650, 700, 750, 800, 850]

#设置切分后对应标签
listLabels = ['300~499','500~549','550~599','600~649','650~699','700~749','750~799','800~850']

#利用pd.cut进行数据离散化切分
data['FICO Group'] = pd.cut(data.cd_score, bins=listBins, labels=listLabels, include_lowest=True,right=False)
```

```python
data["settle_dt"]=pd.to_datetime(data["settle_dt"],format="%d/%m/%Y",errors="coerce")
data["due_dt"]=pd.to_datetime(data["due_dt"],format="%d/%m/%Y",errors="coerce")
data["was_ovdu"]=data.settle_dt>data.due_dt
```

```python
A=data.groupby("FICO Group")["was_ovdu"].sum()/data.groupby("FICO Group")["is_settle"].count()
A.plot(kind="bar")
```

## 2业务维度分析

```python
#不良贷款率随记录时间变化趋势
data["apply_dt"]=pd.to_datetime(data["apply_dt"],format="%d/%m/%Y",errors="coerce")
data["apply_dt_week"]=data.apply_dt.dt.isocalendar().week
```

```python
b=data.groupby("apply_dt_week")['was_ovdu'].sum()/data.groupby("apply_dt_week")["is_settle"].count()
b.plot()
```

```python
data["disbr_dt"]=pd.to_datetime(data["disbr_dt"],format="%d/%m/%Y",errors="coerce")
```

```python
data['loan_term']=data["due_dt"]-data["disbr_dt"]
```

```python
c=data.groupby("loan_term")['was_ovdu'].sum()/data.groupby("loan_term")["is_settle"].count()
c.plot(kind="bar")
```
